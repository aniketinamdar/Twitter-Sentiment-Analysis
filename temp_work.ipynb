```python
def generateTemp(df, category):
    df.reset_index(drop=True, inplace=True)
    temp_df = pd.DataFrame(columns=['text', 'hate_speech', 'cyberbullying', 'crime/threat'])
    temp_df['text'] = df['text']
    if (category == 'hate_speech'):
        temp_df['hate_speech'] = 1
        temp_df['crime/threat'] = 0
        temp_df['cyberbullying'] = 0
    elif (category == 'crime/threat'):
        temp_df['hate_speech'] = 0
        temp_df['crime/threat'] = 1
        temp_df['cyberbullying'] = 0
    elif (category == 'cyberbullying'):
        temp_df['hate_speech'] = 0
        temp_df['crime/threat'] = 0
        temp_df['cyberbullying'] = 1
    return temp_df
```

```python
temp_df = generateTemp(offensive5_hate, 'hate')
categorizer_dataset = categorizer_dataset.append(temp_df, ignore_index = True)
```
